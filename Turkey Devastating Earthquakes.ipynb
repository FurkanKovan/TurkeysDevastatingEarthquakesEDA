## Preprocessing

#### Summary
- Mainly changed missed assigned values.
- For Magnitude (Mag.) column, because 'Mw.' and 'Ms.' metrics are considered approximately similar, taking their raw values is better.
- The hours values splitted from 'Time' column to see their distribution. They rounded up if the time was past half (30'), rounded down if it was before half (30') respectively.
- Ignored 10 missing values from 'Fault Mechanism' field. It's considered unsignificant for our visualization.

```python
prep_data = raw_data.copy()
```

```python
# Turning 'Casualty = 1000+' to '1000' at index 28 for convenience
prep_data.replace('1000+', '1000', inplace=True)
prep_data['Casualty'] = prep_data['Casualty'].astype('int32')
# Turning Mag = '6,1' to '6.1' at index 40 for convenience
prep_data.replace('6,1 (Mw)', '6.1 (Mw)', inplace=True)

# Extracting raw values from Mag. because magnitude (Ms) and (Mw) values are similar
raw_mag = [mag.split(' ')[0] for mag in prep_data['Mag.']]
prep_data['Mag. raw'] = raw_mag

# Splitting hours from time. For convenience rounding up if it's past half and rounding down if it's not
for index, row in prep_data.iterrows():
    if int(row['Time'].split(':')[1]) <= 30:
        prep_data.at[index, 'Time (hours)'] = row['Time'].split(':')[0]
    else:
        prep_data.at[index, 'Time (hours)'] = int(row['Time'].split(':')[0]) + 1
        if prep_data.at[index, 'Time (hours)'] == 24.:
            prep_data.at[index, 'Time (hours)'] = 0.

# Reordering columns
column_names_order = ['Date', 'Time', 'Time (hours)', 'Location', 'Lat.', 'Long.', 'Casualty', 'Mag.', 'Mag. raw', 
                      'Focal Depth (km)', 'Fault Mechanism']
prep_data = prep_data[column_names_order]
```

#### Causalty rates per certain columns

```python
# Printing Casualty rate per relevant column
columns_to_print = ['Date', 'Time (hours)', 'Mag. raw', 'Focal Depth (km)', 'Fault Mechanism']
for col in columns_to_print:
    print('\n' + '-'*40)
    print(prep_data['Casualty'].groupby(prep_data[col]).sum())
```

## Visualization

Executed simple visualizations considering casualty rates for relevant columns. Note that most of these relationships are not explanatory, if not any, because the casualty rate depends on many factors that are not avaliable as data for this given dataset such as durability of buildings, age of the buildings, quality of the materials that used in construction, availability of first aid teams, weather conditions etc. Not to mention the kind of soil the buildings built on.

Cities did not mentioned in these graphs. Considering the different dates it's not easy to interpret these results but feel free to deep further into these findings and share with me if any interesting trends catch your eye!

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.plotting.scatter_matrix(prep_data, figsize=(16,10), alpha=1, diagonal='hist', s=100);
```

```python
columns_to_plot = ['Date', 'Time (hours)', 'Mag. raw', 'Focal Depth (km)', 'Fault Mechanism']

for column in columns_to_plot:
    plt.figure(figsize=(10,6))
    plt.title(f"Casualty vs {column}")
    sns.barplot(x=column, y="Casualty", data=prep_data, capsize=0.2)
    plt.xticks(rotation=90)
```

```python
regplot_columns = ['Date', 'Mag. raw', 'Focal Depth (km)', 'Casualty']
regplot_data = prep_data[regplot_columns]
regplot_data.loc[:, 'Mag. raw'] = regplot_data['Mag. raw'].astype('float')

for column in regplot_columns[:-1]:
    plt.figure(figsize=(8,4))
    plt.title(f"Casualty vs {column}")
    sns.regplot(x=column, y="Casualty", data=regplot_data)
    plt.xticks(rotation=90)
```
